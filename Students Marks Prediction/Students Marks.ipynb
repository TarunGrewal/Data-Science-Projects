```python
# importing libraries and students dataset
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
ds = pd.read_csv('student-mat.csv',delimiter=';')
ds=ds[[ 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic','age', 'Medu', 'Fedu',
        'traveltime', 'studytime',
       'failures', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences', 'G1', 'G2','G3',]]
ds['G3']=ds['G3']>9

ds.info()
```

```python
X = ds.iloc[:, :31].values
y = ds.iloc[:, 31].values
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
for i in range(0,16):
    labelencoder_X_i = LabelEncoder()
    X[:, i] = labelencoder_X_i.fit_transform(X[:, i])
onehotencoder = OneHotEncoder(categorical_features = [6,7,8,9])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 0:]

```

```python
# Splitting the data into Test and Training set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
import statsmodels.api as sm

model = sm.OLS(y_train, X_train).fit()
predictions = model.predict(X_test) 
model.summary()
```

```python

# Scaling the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Dimensionality Reduction using Kernel Principal Component Analysis
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components = 15, kernel = 'rbf')
X_train = kpca.fit_transform(X_train)
X_test = kpca.transform(X_test)

```

```python
# Implementing the ANN model With KERAS using Tensorflow backend
from keras.models import Sequential
from keras.layers import Dense

cl = Sequential()

cl.add(Dense(output_dim =8, init = 'uniform', activation = 'relu', input_dim = 15))

cl.add(Dense(output_dim = 8, init = 'uniform', activation = 'relu'))

cl.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

cl.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

cl.fit(X_train, y_train, batch_size = 10, nb_epoch = 300)
```

```python
# Implementing the ANN model With KERAS using Tensorflow backend
from keras.models import Sequential
from keras.layers import Dense

cl = Sequential()

cl.add(Dense(output_dim =15, init = 'uniform', activation = 'relu', input_dim = 38))

cl.add(Dense(output_dim = 15, init = 'uniform', activation = 'relu'))

cl.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

cl.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

cl.fit(X_train, y_train, batch_size = 10, nb_epoch = 300)
```

```python
#Prediction and Confusion Matrix
y_pred = cl.predict(X_test)
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
